```python
from fastapi import FastAPI

from src.datamodels import TickerRequest

app = FastAPI()

@app.get("/")
def get_ticker(
    request: TickerRequest
):
    df = fetch_stock_data(
        request.ticker,
        days=request.days,
        end_date=request.end_date
    )

    fig = plot_stock_analysis(
        df,
        show_volume=True,
        show_yield=True,
    )

    return {
        'html': fig.to_html(),
    }
```
