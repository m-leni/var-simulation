```python
import yfinance as yf
import pandas as pd

def get_snp_market_caps(sp: pd.DataFrame) -> pd.DataFrame:
    """
    Get market capitalizations for S&P 500 companies.

    Args:
        sp (pd.DataFrame): DataFrame containing S&P 500 tickers.

    Returns:
        pd.DataFrame: DataFrame with tickers and their market capitalizations.
    """
    
    df2 = sp[sp['Asset Type'] == 'Common Stock']
    all = yf.Tickers(' '.join(df2['Ticker'].tolist()))
    marketcaps = {ticker: info.fast_info.get('marketCap', None) for ticker, info in all.tickers.items()}

    return pd.DataFrame.from_dict(marketcaps,orient='index', columns=['Market Cap']).reset_index().rename(columns={'index':'Ticker'})
```
